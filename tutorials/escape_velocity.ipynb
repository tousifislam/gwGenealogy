```python
vesc_gc = gwGenealogy.vescape.sample_globular_cluster_vesc_distribution(n_samples=1000)
vesc_etg = gwGenealogy.vescape.sample_early_type_galaxies_vesc_distribution(n_samples=1000)
```

```python
plt.figure(figsize=(6,4))
plt.hist(vesc_gc, bins=20, histtype='stepfilled', density=True, alpha=0.4, label='$\\rm GCs$');
plt.xlabel('$v_{\\rm esc}$', fontsize=16)
plt.ylabel('$p(v_{\\rm esc})$', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(vesc_gc, bins=20, histtype='stepfilled', alpha=0.4, 
         label='GCs', edgecolor='none', density=False, 
         weights=np.ones(len(vesc_gc))/np.max(np.histogram(vesc_gc, bins=20)[0]))
plt.hist(vesc_etg, bins=20, histtype='stepfilled', alpha=0.4, 
         label='ETGs', edgecolor='none', density=False, 
         weights=np.ones(len(vesc_etg))/np.max(np.histogram(vesc_etg, bins=20)[0]))
plt.xlabel('$v_{\\rm esc}$', fontsize=16)
plt.ylabel('$p(v_{\\rm esc})$', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
