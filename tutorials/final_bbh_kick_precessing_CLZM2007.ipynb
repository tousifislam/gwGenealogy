```python
def test_bbh_final_kick_precessing_CLZM2007():
    """Test the kick calculator with various binary configurations"""
    
    print("************* Testing Binary Black Hole Kick Velocities *************")
    
    # Test cases covering different scenarios
    test_cases = [
        # Case 1: Equal mass, aligned spins
        {"name": "Equal mass, aligned spins", "q": 1.0, "chi1": 0.9, "chi2": 0.9, 
         "theta1": 0, "theta2": 0, "phi12": 0},
        
        # Case 2: Equal mass, anti-aligned spins  
        {"name": "Equal mass, anti-aligned spins", "q": 1.0, "chi1": 0.9, "chi2": 0.9,
         "theta1": 0, "theta2": 180, "phi12": 0},
        
        # Case 3: Equal mass, perpendicular spins
        {"name": "Equal mass, perpendicular spins", "q": 1.0, "chi1": 0.9, "chi2": 0.9,
         "theta1": 90, "theta2": 90, "phi12": 0},
        
        # Case 4: Unequal mass, high spins
        {"name": "Unequal mass (q=0.3), high spins", "q": 0.3, "chi1": 0.95, "chi2": 0.95,
         "theta1": 60, "theta2": 120, "phi12": 90},
        
        # Case 5: Extreme mass ratio
        {"name": "Extreme mass ratio (q=0.1)", "q": 0.1, "chi1": 0.8, "chi2": 0.5,
         "theta1": 45, "theta2": 135, "phi12": 180},
        
        # Case 6: High spinning, misaligned
        {"name": "High spins, misaligned", "q": 0.8, "chi1": 0.98, "chi2": 0.95,
         "theta1": 30, "theta2": 150, "phi12": 45},
        
        # Case 7: Superkick configuration (should give high kicks)
        {"name": "Superkick configuration", "q": 1.0, "chi1": 0.8, "chi2": 0.8,
         "theta1": 90, "theta2": 90, "phi12": 180},
    ]
    
    max_kick_found = 0
    max_kick_config = None
    
    for i, case in enumerate(test_cases):
        print(f"\nCase {i+1}: {case['name']}")
        print(f"Parameters: q={case['q']}, χ₁={case['chi1']}, χ₂={case['chi2']}")
        print(f"           θ₁={case['theta1']}°, θ₂={case['theta2']}°, φ₁₂={case['phi12']}°")
        
        # Test with multiple random Theta values
        kicks = []
        for _ in range(100):  # Sample 100 random Theta values
            kick, vm, vs_perp, vs_par, theta = bbh_final_kick_precessing_CLZM2007(
                case['q'], case['chi1'], case['chi2'], 
                np.radians(case['theta1']), np.radians(case['theta2']), np.radians(case['phi12']),
                debug=True)
            kicks.append(kick)
        
        kicks = np.array(kicks)
        max_kick = np.max(kicks)
        mean_kick = np.mean(kicks)
        
        print(f"Kick velocities: Max = {max_kick:.1f} km/s, Mean = {mean_kick:.1f} km/s")
        
        if max_kick > max_kick_found:
            max_kick_found = max_kick
            max_kick_config = case
        
        # Show detailed breakdown for one calculation
        kick, vm, vs_perp, vs_par, theta = bbh_final_kick_precessing_CLZM2007(
            case['q'], case['chi1'], case['chi2'], 
            np.radians(case['theta1']), np.radians(case['theta2']), np.radians(case['phi12']),
            debug=True)
        
        print(f"Components: V_m = {vm:.1f}, V_s⊥ = {vs_perp:.1f}, V_s∥ = {vs_par:.1f} km/s")
    
    print(f"MAXIMUM KICK FOUND: {max_kick_found:.1f} km/s")
    print(f"Configuration: {max_kick_config['name']}")
    print(f"Expected range: ~5000 km/s (from literature)")
    print("************* Done! *************")
    
    return max_kick_found, max_kick_config
```

```python
def find_maximum_bbh_final_kick_precessing_CLZM2007():
    """Search for configurations that produce maximum kicks"""
    
    print("************* Searching for maximum kick configurations *************")
    
    max_kick = 0
    best_config = None
    
    # Grid search over parameter space
    q_values = [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]
    chi_values = [0.0, 0.5, 0.8, 0.9, 0.95, 0.99]
    theta_values = np.radians([0, 30, 45, 60, 90, 120, 135, 150, 180])
    phi_values = np.radians([0, 45, 90, 135, 180])
    
    total_configs = 0
    high_kick_configs = []
    
    for q in q_values:
        for chi1 in chi_values:
            for chi2 in chi_values:
                for theta1 in theta_values:
                    for theta2 in theta_values:
                        for phi12 in phi_values:
                            total_configs += 1
                            
                            # Test with multiple Theta values
                            kicks = []
                            for _ in range(10):
                                kick, _, _, _, _ = bbh_final_kick_precessing_CLZM2007(
                                    q, chi1, chi2, theta1, theta2, phi12, debug=True)
                                kicks.append(kick)
                            
                            max_kick_this_config = np.max(kicks)
                            
                            if max_kick_this_config > max_kick:
                                max_kick = max_kick_this_config
                                best_config = {
                                    'q': q, 'chi1': chi1, 'chi2': chi2,
                                    'theta1': theta1, 'theta2': theta2, 'phi12': phi12,
                                    'kick': max_kick_this_config
                                }
                            
                            if max_kick_this_config > 3000:  # High kick threshold
                                high_kick_configs.append({
                                    'q': q, 'chi1': chi1, 'chi2': chi2,
                                    'theta1': theta1, 'theta2': theta2, 'phi12': phi12,
                                    'kick': max_kick_this_config
                                })
    
    print(f"Searched {total_configs} configurations")
    print(f"Found {len(high_kick_configs)} with kicks > 3000 km/s")
    print(f"\nAbsolute maximum kick: {max_kick:.1f} km/s")
    print("Best configuration:")
    for key, value in best_config.items():
        if key != 'kick':
            print(f"  {key}: {value}")
    
    # Show top 10 high-kick configurations
    high_kick_configs.sort(key=lambda x: x['kick'], reverse=True)
    print(f"\nTop 10 highest kick configurations:")
    for i, config in enumerate(high_kick_configs[:10]):
        print(f"{i+1:2d}. {config['kick']:6.1f} km/s - "
              f"q={config['q']}, χ₁={config['chi1']}, χ₂={config['chi2']}, "
              f"θ₁={config['theta1']}, θ₂={config['theta2']}, φ₁₂={config['phi12']}")

    print("************* Done! *************")
    
    return max_kick, best_config, high_kick_configs
```
